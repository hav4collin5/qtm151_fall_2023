<font size = "5">

(a) Import and parse dates

<font size = "3">

- Import the dataset "currencies.csv"
- Parse the "date_string" column with ```pd.to_datetime()```
- To get full points, set the "format" option using wildcards as in Lecture 20.


````python
# Write your own code
file_path = '/Users/havacollins/Desktop/qtm151_fall_2023/quizzes/quiz_5/data_raw/currencies.csv'
data = pd.read_csv(file_path)

data['date_string'] = pd.to_datetime(data['date_string'], format='%A, %b %d, %Y')

# Extract month and week number as strings
data['month_str'] = data['date_string'].dt.strftime("%m")  
data['week_str'] = data['date_string'].dt.strftime("%W")   
````

<font size = "5">

(b) Plot multiple time series

<font size = "3">

- Create a line plot plot which has the "gbp_to_usd" and "eur_to_usd" on the y-axis <br>
and the date on the x-axis
- Label the x-axis, the y-axis, and the title
- Add a legend and set the names to "1 Euro to USD" and "1 GPB to USD"

Hint: See Lecture 21 for how to plot multiple columns <br>
Hint: Remember to include the legend names as a list ```[...]```


````python
# Write your own code
plt.figure(figsize=(10, 6))  
plt.plot(data['gbp_to_usd'], label='1 GBP to USD')  
plt.plot(data['eur_to_usd'], label='1 EUR to USD') 

# Labeling
plt.xlabel('Date')  
plt.ylabel('Exchange Rate')  
plt.title('GBP and EUR to USD Exchange Rates Over Time')  

plt.legend()

plt.show()
````

<font size = "5">

(c) Compute growth rates

<font size = "3">

- Calculate a new column "growth_euro" which computes the percentage daily <br>
growth rate of "eur_to_usd".

Hint: Use ```.diff()``` and ```.shift(1)``` as in Lecture 21

````python
# Write your own code

data['growth_euro'] = data['eur_to_usd'].diff() / data['eur_to_usd'].shift(1) * 100
````

<font size = "5">

(d) Flag subset of data and plot

<font size = "3">

- Create a boolean colum (True/False) "bool_cond" which checks whether <br>
"growth_euro" had a value greater than 2 or less than -2
- Generate a line plot with the "growth_euro" series, label the axes and the title
- Use the command ```plt.fill_between()``` as in Lecture 21 to create a shaded region <br>
for time periods where "bool_cond" is true.



````python
# Write your own code
data['bool_cond'] = (data['growth_euro'] > 2) | (data['growth_euro'] < -2)

plt.figure(figsize=(10, 6))  # Adjust the size as needed
plt.plot(data['growth_euro'], label='Euro Growth Rate (%)')  

plt.fill_between(data.index, data['growth_euro'], where=data['bool_cond'], color='gray', alpha=0.5)

plt.xlabel('Date')  # X-axis Label
plt.ylabel('Growth Rate (%)')  # Y-axis Label
plt.title('Daily Euro to USD Growth Rate with Significant Changes Highlighted')  
plt.legend()

plt.show()
````

<font size = "5">

(e) Aggregate dataset by time period

<font size = "3">

- Create an object "month_config" that uses ```pd.Grouper()```, setting the key to <br>
the column "date" and freq to "m" (month)
- Follow the instructions in Lecture 20 to compute an aggregate dataset <br>
by month which computes the standard deviation of "eur_to_usd"
- Display the dataset




````python
# Write your own code
month_grouper = pd.Grouper(freq='M')

monthly_std_dev = data.groupby(month_grouper).std()['eur_to_usd']

print(monthly_std_dev)
````
